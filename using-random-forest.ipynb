```python
train_df.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis = 1, inplace = True)
test_df.drop(['PassengerId','Name','Ticket','Cabin','Embarked'], axis = 1, inplace = True)
```

```python
train_df.fillna(value = train_df.median(),inplace = True)
```

```python
test_df['Fare'].fillna(value = t,inplace = True)
```

```python
test_df.fillna(value = test_df.mean(),inplace = True)
```

```python
train_df.describe()
```

```python
m = train_df.drop(['Survived','Sex'],axis = 1).max() / 2
m
```

```python
train_df.Fare = (train_df['Fare'] - m['Fare'])/train_df['Fare'].max()
```

```python
#train_df.SibSp = (train_df['SibSp'] - m['SibSp'])/train_df['SibSp'].max()
train_df.Age = (train_df['Age'] - m['Age'])/train_df['Age'].max()
#train_df.Parch = (train_df['Parch'] - m['Parch'])/train_df['Parch'].max()
```

```python
test_df.Fare = (test_df['Fare'] - m['Fare'])/test_df['Fare'].max()
#test_df.SibSp = (test_df['SibSp'] - m['SibSp'])/test_df['SibSp'].max()
test_df.Age = (test_df['Age'] - m['Age'])/test_df['Age'].max()
#test_df.Parch = (test_df['Parch'] - m['Parch'])/test_df['Parch'].max()
```

```python
train_df.head()
```

```python
gender_map = {
    'male' : 0,
    'female': 1
}

train_df.Sex = train_df.Sex.map(gender_map)
test_df.Sex = test_df.Sex.map(gender_map)
```

```python
test_df[test_df.Parch == 9]
```

```python
features = ["Pclass","SibSp", "Parch"]
for i in features :
    dummies = pd.get_dummies(train_df[i])
    train_df = pd.concat([train_df,dummies],axis = 1)
    train_df.drop(i,inplace = True,axis = 1)
    dummies = pd.get_dummies(test_df[i])
    test_df = pd.concat([test_df,dummies],axis = 1)
    test_df.drop(i,inplace = True,axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(train_df.drop('Survived',axis = 1),train_df['Survived'],test_size = 0.15,random_state = 45)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 1750,max_depth=10,
                                           min_samples_split=6,
                                           min_samples_leaf=6,
                                           max_features='auto')
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
predict = clf.predict(test_df)
```

```python
passengerid = pd.read_csv('/kaggle/input/titanic/test.csv').PassengerId
Submission = pd.DataFrame({'PassengerId':passengerid,'Survived': predict})
Submission.head(10)
```
