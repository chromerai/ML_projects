```python
gender_map = {
    'male' : 0,
    'female' : 1
}

train_df.Sex = train_df.Sex.map(gender_map)
train_df.head()
```

```python
test_df.Sex = test_df.Sex.map(gender_map)
test_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_df.drop('Survived',axis = 1),train_df['Survived'],test_size = 0.15,random_state = 15)
#print(X_train.shape,y_train.shape)
#print(X_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = 'liblinear',multi_class = 'ovr')
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
print(lr.coef_)
```

```python
predict = lr.predict(test_df)
predict[:10]
```

```python
passengerid = pd.read_csv('/kaggle/input/titanic/test.csv').PassengerId
Submission = pd.DataFrame({'PassengerId':passengerid,'Survived': predict})
Submission.head(10)
```
